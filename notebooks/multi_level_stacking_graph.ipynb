```python
import Orange
import matplotlib.pyplot as plt
import pandas as pd

# path = 'C:/Users/cantao/Documents/Python scripts/Critical Difference Diagrams/fabiana/dados.csv'
# data = pd.read_csv(path)
# print(data)
data = pd.DataFrame(x)

#gets the mean for each column ranks
data.rank(axis='columns',method='average',ascending=False).mean()
```

```python
names = data.columns
average_ranks = list()
for item in data.rank(axis='columns',method='average',ascending=False).mean():
    average_ranks.append(item)
names = ["C", "SC 5", "SC 500", "SC 1000", "MPC 5", "MPC 500", "MPC 1000"]
# 'cd' finds the critical difference range
cd = Orange.evaluation.compute_CD(average_ranks, 5, alpha='0.05', test='bonferroni-dunn')  #bonferroni-dunn   #nemenyi
Orange.evaluation.graph_ranks(average_ranks, names, cd=cd, width=3, textspace=0.1)

# plt.savefig('CD.pdf', bbox_inches='tight',pad_inches = 0)
plt.show()
```
